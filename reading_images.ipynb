Images are represented as array of numbers denoting the pixels they represent, these arrays are very large depending on the size of the image. for a multiple color image the images matrix is made of other matrices denoting the color for each matrix of the images

They are several libararies one can use for image processing, scipy.ndimage, numpy and last but not least opencv. opencv being our focus for the is lession is a library built on top of numpy. it has several methods for dealing with images and videos (which are basically image frames played over time).

when processing a images the first thing you need to do is to read the image, to carry out this functionality in opencv we do the following

Note: make you have opencv installed on your platform of choice either windows, linux or mac

```python
import cv2 # the name opencv is reffered to in python

img = cv2.imread('Images/autumn-season-leafs-plant-scene-generative-ai.jpg') # reads the image and stores it in the img variable

```

While reading an image there are filter we can apply to it to make it easier to process 

cv2.IMREAD_GRAYSCALE - when this filter is passed to the image read function (cv2.imread()) it removes the alpha channel which indicates the level of opacity of the images and converts the image to a gray scale. which makes it easier for processing because that one matrix less that we have to process

cv2.IMREAD_COLOR - this filter tells the image read function to read the image with its full color channels
cv2.IMREAD_UNCHANGED - this filter tells the image read function to read the image as is "unchanged"

they can be abbreviated as
cv2.IMREAD_UNCHANGED = -1
cv2.IMREAD_COLOR = 1
cv2.IMREAD_GRAYSCALE = 0

To read and display an image we use the following code

```python
resized_img = cv2.resize(img, (200,200))
cv2.imshow('<The title of the image>', resized_img) 
cv2.waitKey(0) # waits for any key press
cv2.destroyAllWindows() # closes the window of display
```
